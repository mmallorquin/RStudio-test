```
# Iterate over the sequence of column names
res = list()
for column in df:
   # Select column contents by column name using [] operator
   count = 0
   columnObj = df[column]
      
   temp_res = {
       'Movie': column,
       'People rated': columnObj.notnull().sum(),
   }
   res.append(temp_res)
   
```
